```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel("0.xls")
plt.scatter(df.X, df.Y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid()

x = np.linspace(df["X"].min(), df["X"].max(), 100)
y = (8/9) * x + 5

plt.plot(x, y)
plt.show()

a0 = []
a1 = []
for i in df.values:
    x = i[0]
    y = i[1]
    v = y - ((8/9) * x + 5)
    if v < 0:
        a0.append((x, y))
    else:
        a1.append((x, y))
xs = []
ys = []
for clast in [a0, a1]:
    mx = 10 ** 50
    x = -1
    y = -1
    for p in clast:
        val = 0
        for q in clast:
            val += ((q[0] - p[0]) * (q[0] - p[0]) + (q[1] - p[1]) * (q[1] - p[1])) ** 0.5
        if val < mx:
            best = val
            x = p[0]
            y = p[1]
    xs.append(x)
    ys.append(y)
    plt.scatter([q[0] for q in clast], [q[1] for q in clast])
print(int(sum(xs) / len(xs) * 1000), int(sum(ys) / len(ys) * 1000))
```
